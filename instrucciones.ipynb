1. DEFINICIÓN DEL PROBLEMA (compartida por todos)
Escenario concreto a abordar:
La desinformación en temas de salud (productos “milagro”, curas falsas, antivacunas) se propaga mediante sitios de baja credibilidad que imitan medios de noticias, apoyados en contenido generado por LLMs. Un ciudadano no especialista no puede distinguir si una afirmación alarmante está verificada o proviene de una fuente sin rigor periodístico.

Enfoque del prototipo:
Crear un sistema automatizado que recolecte fact-checks recientes sobre salud desde Snopes, extraiga los dominios que originaron el rumor, y los cruce con la base de datos de Media Bias/Fact Check (MBFC) para determinar el nivel de factualidad de dichos dominios. El resultado es una “alerta temprana” que muestra qué narrativas falsas están siendo impulsadas por fuentes de muy baja credibilidad.

5. DETALLE DE CADA FASE (lo que debe contener cada paso)
5.1 Elección y justificación de fuentes (Integrante A con insumos de B y C)
Snopes:

robots.txt: /fact-check/ está permitida. No se necesita autenticación.

Estructura: cada artículo tiene una etiqueta article con enlace a /fact-check/.... En la página de detalle, div.claim, div.rating-name, time.date, y enlace externo al rumor.

Variables extraíbles: claim, rating, fecha, categoría (tags), dominio de la fuente original.

MBFC:

robots.txt: no bloquea el directorio raíz ni las categorías. Se permite scraping con User-Agent identificable.

Listados: https://mediabiasfactcheck.com/category/fake-news/, paginados con /page/2/.

Detalle: nombre del medio, enlace a su web, bias (izquierda, conspiración, etc.), factual reporting (HIGH, MIXED, LOW, VERY LOW).

Variables: domain (extraído del enlace), bias, factual_reporting.

Criterio ético: Ambas son webs de acceso público, no se extraen datos personales, no se sobrecarga el servidor (caché + delays), y se respeta robots.txt. En el informe se incluirán capturas del robots.txt de cada sitio.

5.2 Estrategia de web scraping (detalle para B y C)
Tecnologías comunes:

requests + BeautifulSoup (no se necesita Selenium, el contenido es estático).

requests-cache con backend SQLite, expiración 24h.

Rotación de 8 User-Agents reales (se pueden copiar de https://user-agents.net/).

time.sleep(random.uniform(2, 5)) entre peticiones al mismo dominio.

Mecanismo de reintentos con requests.adapters.HTTPAdapter (max_retries=3).

Manejo de paginación:

Snopes: en la página de listado, buscar a.next.page-numbers para obtener la URL siguiente. Si no existe, terminar.

MBFC: similar, con a.next.page-numbers o link[rel="next"].

Manejo de errores:

Si un artículo no tiene alguno de los campos, se deja vacío y se registra en un log.

Si el servidor devuelve 403, se para el scraper y se notifica para revisar bloqueos. En entorno real se rotarían proxies, pero para el prototipo basta con simular con datos pre-descargados si ocurre.

Estructura de los scripts:

config.py: constantes (URLs base, lista de User-Agents, delay, etc.)

utils.py: función get_page(url) que usa sesión cacheada, elige User-Agent y hace sleep.

snopes_scraper.py: extract_list_page(url), extract_detail(article_url), main() que itera y guarda CSV.

mbfc_scraper.py: análogo.

5.3 Arquitectura de la solución (Responsable A)
Pipeline de datos:

text
[Snopes Scraper] ----> fact_checks.csv -----|
                                          |----> [Integración] ----> report.csv
[MBFC Scraper]   ----> sources.csv   ------|
Opcional: volcado intermedio a SQLite para facilitar consultas.

Lógica de integración:

Cargar fact_checks.csv y sources.csv en DataFrames de pandas.

Normalizar dominios (eliminar www., extraer solo dominio.tld con tldextract).

Para cada fact-check con origin_domain no nulo, buscar una coincidencia exacta o contenida en sources['domain']. (Ej.: si el rumor salió en www.badnews.com y en MBFC está badnews.com, se asocian).

Almacenar la unión con los campos de ambas fuentes.

Exportar a integrated_report.csv.

Interfaz de salida:

Terminal: imprime resumen con print de los top 10 dominios de baja factualidad que más fact-checks falsos acumulan.

Archivo CSV para visualización en Excel.

Actualización:
Por lotes (batch) semanal. No se requiere tiempo real.

5.4 Validación y métricas (Responsable A)
Métrica cuantitativa principal:
Porcentaje de fact-checks con rating "False" (o "Mostly False") cuyo dominio de origen tiene factualidad "Low" o "Very Low".

Se calcula: (número de claims falsos con dominio de baja calidad) / (total de claims falsos con dominio identificado) * 100.

Meta deseada: que al menos el 60% de los bulos provengan de fuentes de baja credibilidad, lo que validaría que el sistema identifica focos de desinformación.

Otras métricas:

Cobertura: nº de fact-checks descargados (>50).

Precisión del cruce: revisar manualmente 20 registros y contar cuántos dominios fueron correctamente emparejados (se espera >80%).

Completitud: porcentaje de artículos donde se pudo extraer origin_domain (en Snopes no siempre se menciona).

Validación manual:
Se elegirán 3 ejemplos concretos, se capturará pantalla del artículo en Snopes y la ficha en MBFC, y se discutirá por qué el sistema ayudó a entender la desinformación. Esto va como evidencia en el informe.

Limitaciones:

Cambios en el HTML rompen los scrapers; se mitigará con un sistema de alertas (asserción de campos obligatorios).

Escalabilidad: para millones de registros se necesitaría una cola de trabajo y proxies rotativos; fuera del alcance.

Sesgo idiomático: ambas fuentes son en inglés, por lo que la desinformación en español no se cubre (se discute como trabajo futuro).

5.5 Consideraciones éticas y legales (Responsable D)
El capítulo ético cubrirá:

Respeto a robots.txt: mostramos captura de las secciones permitidas y explicamos que solo accedemos a rutas públicas no restringidas.

No violación de ToS: ambos sitios permiten investigación académica no comercial; nuestro uso es exclusivamente educativo. No se extraen ni publican textos completos que puedan infringir copyright, solo metadatos.

Protección de datos personales: no se recolecta información de usuarios ni comentarios, solo datos de los propios artículos (hechos públicos). No hay scraping de perfiles.

Identificación del bot: el User-Agent incluye un string explicativo EstudiantesUP/1.0 (se puede cambiar por el nombre de la universidad). Se podría contactar a los administradores para notificar la actividad, pero dado que el tráfico es mínimo (< 200 peticiones en una semana) no es necesario.

Posibilidad de pedir permiso: en un entorno real se solicitaría autorización o se usarían APIs oficiales si existieran. Como no es el caso, nuestro scraping no intrusivo es asimilable a la actividad de un navegador.
